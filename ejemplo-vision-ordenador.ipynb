```python
from scipy import misc
face = misc.face() # Get a 1024 x 768, color image of a raccoon face.
misc.imsave('face.png', face) # First we need to create the PNG file. Save an array as an image.

face = misc.imread('face.png')
type(face)      

face.shape, face.dtype
```

```python
face.tofile('face.raw') # Create raw file
face_from_raw = np.fromfile('face.raw', dtype=np.uint8)
face_from_raw.shape

face_from_raw.shape = (768, 1024, 3)
```
