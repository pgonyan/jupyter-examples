```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.charts import Bar
from bokeh.charts.attributes import ColorAttr, CatAttr
#from bokeh.charts.builders.bar_builder import BarBuilder
import bokeh
print (bokeh.__version__)

dict = {
    'categorias': ['rojos','avez','azules','azules'],
    'valores': [10 ,2,3,6666],
    'valores2': [102 ,22,3,666]
    }
df2 = pd.DataFrame(dict)

df2

```

```python
label = CatAttr(df=df2, columns='categorias', sort=False)

# p2 = Bar(df2, 'categorias',values='valores2', title="test chart",bar_width=0.2,legend=None)
p2 = Bar(df2, label=label,values='valores2', title="Ejemplos",bar_width=0.2,legend=None)
 
output_notebook()

show(p2)
```
